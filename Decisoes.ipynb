# **Identificação de caracteres**

**RADOM FOREST - "MNIST"**

**1. Importando bibliotecas convenientes**

```
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

**2. Carregando dataframe de estudo**

```
numeros = load_digits()
df = pd.DataFrame(numeros.data)
df['alvo'] = numeros.target
df
```

**3. Dividindo dataset**

```
#Identificando quem é dado e alvo
x = numeros.data
y = numeros.target
#Dividindo "dados" de treino/teste
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.2)
#Definindo modelo
modelo = RandomForestClassifier(n_estimators=20)
#Executando treino
modelo.fit(x_treino,y_treino)
#Avaliando acurácia
print(modelo.score(x_teste,y_teste))
```

**4. Identificando frequências de classificação para cada classe do modelo**

```
#Projetando matriz
y_predito = modelo.predict(x_teste)
matriz_confusao = confusion_matrix(y_teste,y_predito)
#Heatmap
plt.figure(figsize=(9,6))
sn.heatmap(matriz_confusao, annot=True,cmap="YlGnBu")
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
plt.show()
```
